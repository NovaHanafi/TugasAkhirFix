```python
X=df.drop(['Unnamed: 0.1.1','Unnamed: 0','Unnamed: 0.1'],axis=1)
X.drop(['SimillarHTTP','Flow ID',' Source IP', ' Destination IP', ' Timestamp'],axis=1,inplace=True)
```

```python
X.info()
```

```python
X.replace([np.inf, -np.inf], np.nan, inplace=True)
X.dropna(inplace=True)
```

```python
y=X[' Label']
X.drop([' Label'],axis=1,inplace=True)
```

```python
y.value_counts().sum()
```

```python
X.head(5)
```

```python
X.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
scaler.transform(X)
```

```python
X.head()
```

```python
from sklearn.feature_selection import chi2
chi=chi2(scaler.transform(X),y)
```

```python
chi
```

```python
Xcol=pd.DataFrame(X.columns)
Xcol
```

```python
chi2=pd.DataFrame(chi[0])
pval=pd.DataFrame(chi[1])
```

```python
result=pd.concat([Xcol,chi2,pval],axis=1)
result.columns =['Fitur','Chi2','P value']
result.dropna(inplace=True)
```
